# Feature Engineering

```python
df_restaurants['is_rated'] = df_restaurants['rate'].notna()
```

```python
def categorize_rating(r):
    if pd.isna(r):
        return 'Not Rated'
    elif r >= 4.5:
        return 'Excellent'
    elif r >= 4.0:
        return 'Very Good'
    elif r >= 3.5:
        return 'Good'
    elif r >= 3.0:
        return 'Average'
    else:
        return 'Poor'

df_restaurants['rating_category'] = df_restaurants['rate'].apply(categorize_rating)
```

```python
df_restaurants['rating_category'].value_counts()
```

```python
def categorize_cost(c):
    if pd.isna(c):
        return 'Unknown'
    elif c < 300:
        return 'Budget'
    elif c < 700:
        return 'Mid-range'
    elif c < 1500:
        return 'Premium'
    else:
        return 'Luxury'

df_restaurants['cost_category'] = df_restaurants['approx_cost'].apply(categorize_cost)
```

```python
df_restaurants['cost_category'].value_counts()
```

```python
def categorize_votes(v):
    if v == 0:
        return 'No Engagement'
    elif v < 25:
        return 'Low'
    elif v < 120:
        return 'Moderate'
    elif v < 500:
        return 'High'
    else:
        return 'Very High'

df_restaurants['engagement_level'] = df_restaurants['votes'].apply(categorize_votes)
```
